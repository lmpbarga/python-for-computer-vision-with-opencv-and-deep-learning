```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2 as cv

img = cv.imread('DATA/dog_backpack.jpg')
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.rectangle(fix_img,pt1=(200,700),pt2=(600,400),color=(255,0,0),thickness=10)
plt.imshow(fix_img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2 as cv

img = cv.imread('DATA/dog_backpack.jpg')
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
vertices = np.array([[250,700],[450,400],[650,700]],dtype=np.int32)
vertices.shape
pts = vertices.reshape((-1,1,2))
cv.polylines(fix_img,[pts],isClosed=True,color=(0,0,255),thickness=10)
plt.imshow(fix_img)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2 as cv

img = cv.imread('DATA/dog_backpack.jpg')
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
vertices = np.array([[250,700],[450,400],[650,700]],dtype=np.int32)
vertices.shape
pts = vertices.reshape((-1,1,2))
cv.fillPoly(fix_img,[pts],color=(0,0,255))
plt.imshow(fix_img)
```

```python
import numpy as np
import cv2 as cv

img = cv.imread('DATA/dog_backpack.jpg')
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

def draw_circle(event,x,y,flags,param):
    if event == cv.EVENT_LBUTTONDOWN:
        cv.circle(img,(x,y),100,(0,0,255),10)

cv.namedWindow(winname='my_drawing')

cv.setMouseCallback('my_drawing', draw_circle)

while True:
    cv.imshow('my_drawing',img)
    
    if(cv.waitKey(20) & 0xFF == 27):
        break


cv.destroyAllWindows()
```
