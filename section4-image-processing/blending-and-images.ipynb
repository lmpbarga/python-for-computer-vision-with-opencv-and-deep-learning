```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2 as cv

img1 = cv.cvtColor(cv.imread('../DATA/dog_backpack.jpg'), cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(cv.imread('../DATA/watermark_no_copy.png'), cv.COLOR_BGR2RGB)

img1 = cv.resize(img1,(1200,1200))
img2 = cv.resize(img2,(1200,1200))

blended = cv.addWeighted(src1=img1,alpha=0.9,src2=img2,beta=0.1,gamma=0)

plt.imshow(blended)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2 as cv
import numpy as np

img1 = cv.cvtColor(cv.imread('../DATA/dog_backpack.jpg'), cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(cv.imread('../DATA/watermark_no_copy.png'), cv.COLOR_BGR2RGB)

img2 = cv.resize(img2,(600,600))

x_offset = 934-600
y_offset = 1401-600

roi = img1[y_offset:1401,x_offset:943]

img2gray = cv.cvtColor(img2, cv.COLOR_RGB2GRAY)

mask_inv = cv.bitwise_not(img2gray)

white_background = np.full(img2.shape,255,dtype=np.uint8)

bk = cv.bitwise_or(white_background,white_background,mask=mask_inv)

fg = cv.bitwise_or(img2,img2,mask=mask_inv)

final_roi = cv.bitwise_or(roi,fg)

large_img = img1

small_img = final_roi

large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]] = small_img

plt.imshow(large_img)
```
