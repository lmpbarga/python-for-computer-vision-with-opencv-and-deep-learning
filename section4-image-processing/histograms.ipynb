```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def show_img(img_path):
    img = cv.imread(img_path)
    show_img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    return img,show_img

horse,show_horse = show_img('../DATA/horse.jpg')
rainbow,show_rainbow = show_img('../DATA/rainbow.jpg')
bricks,show_bricks = show_img('../DATA/bricks.jpg')
```

```python
plt.imshow(bricks)
```

```python
hist_values = cv.calcHist([bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
hist_values = cv.calcHist([horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
img = bricks
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('HISTOGRAM FOR BRICKS')
```

```python
img = horse
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])
plt.title('HISTOGRAM FOR HORSE')
```

```python
img = rainbow
mask = np.zeros(img.shape[:2],np.uint8)
mask[300:400,100:400] = 255
masked_img = cv.bitwise_and(img,img,mask=mask)
show_masked_img = cv.bitwise_and(show_rainbow,show_rainbow,mask=mask)
hist_mask_values_red = cv.calcHist([rainbow],channels=[2],mask=mask,histSize=[256],ranges=[0,256])
plt.plot(hist_mask_values_red)
plt.title('RED HISTOGRAM FOR MASKED RAINBOW')
```

```python
hist_values_red = cv.calcHist([rainbow],channels=[2],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values_red)
plt.title('RED HISTOGRAM FOR RAINBOW')
```
