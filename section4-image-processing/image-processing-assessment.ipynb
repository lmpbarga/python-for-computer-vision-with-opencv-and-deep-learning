```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
giraffes = cv.imread('../DATA/giraffes.jpg')
fix_giraffes = cv.cvtColor(giraffes, cv.COLOR_BGR2RGB)
display_img(fix_giraffes)
```

```python
giraffes = cv.imread('../DATA/giraffes.jpg',0)
ret,th1 = cv.threshold(giraffes,127,255,cv.THRESH_BINARY)
display_img(th1,cmap='gray')
```

```python
giraffes = cv.imread('../DATA/giraffes.jpg')
hsv_giraffes = cv.cvtColor(giraffes,cv.COLOR_BGR2HSV)
display_img(hsv_giraffes)
```

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
giraffes = cv.imread('../DATA/giraffes.jpg')
fix_giraffes = cv.cvtColor(giraffes,cv.COLOR_BGR2RGB)
result = cv.filter2D(fix_giraffes,-1,kernel)
display_img(result)
```

```python
giraffes = cv.imread('../DATA/giraffes.jpg',0)
sobelx = cv.Sobel(giraffes,cv.CV_64F,1,0,ksize=5)
display_img(sobelx,cmap='gray')
```

```python
giraffes = cv.imread('../DATA/giraffes.jpg')

color = ['b','g', 'r']

for i,col in enumerate(color):
    histr = cv.calcHist([giraffes],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
plt.title('MY HISTO')
```
