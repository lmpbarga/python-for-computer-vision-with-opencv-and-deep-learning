```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)

gorilla = cv.imread('../DATA/gorilla.jpg',0)

display(gorilla,cmap='gray')
```

```python
hist_values = cv.calcHist([gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
eq_gorilla = cv.equalizeHist(gorilla)
display(eq_gorilla,cmap='gray')
```

```python
hist_values = cv.calcHist([eq_gorilla],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_values)
```

```python
gorilla = cv.imread('../DATA/gorilla.jpg')
show_gorilla = cv.cvtColor(gorilla,cv.COLOR_BGR2RGB)
display(show_gorilla)
```

```python
hsv = cv.cvtColor(gorilla,cv.COLOR_BGR2HSV)
hsv[:,:,2].min()
hsv[:,:,2] = cv.equalizeHist(hsv[:,:,2])
eq_gorilla = cv.cvtColor(hsv,cv.COLOR_HSV2RGB)
display(eq_gorilla)
```
